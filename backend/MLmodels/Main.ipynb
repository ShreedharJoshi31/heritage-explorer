Feature Selection Part

```python
tours.columns
```

```python
tours=tours[['id', 'title', 'city', 'desc', 'class']]
```

```python
tours
```

```python
tours['tags'] = tours['desc']+tours['city']+tours['class']
```

```python
tours
```

```python
new_data  = tours.drop(columns=['desc', 'city', 'class'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="Elephanta Caves"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:4]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(tours):
    index=new_data[new_data['title']==tours].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].id)
```

```python
recommand("Raigad Fort")
```
